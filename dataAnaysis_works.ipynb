```python
import pandas as pd

df = pd.read_csv('cleaned_goodreads_works.csv')  # Load the cleaned data
print(df)
```

```python
#filter for books with published year after 2010
df_filtered = df[df['original_publication_year'] > 2013]
print(df_filtered)
```

```python
  # Example analysis: Count the number of works by each author
author_counts = df_filtered['author'].value_counts()
    
   # Print the results
print("Number of Works by Each Author:")
print(author_counts)
    
 # Example analysis: Find the average rating of works
average_rating = df_filtered['avg_rating'].mean()
    
print("\nAverage Rating of Works:", average_rating)
# Example analysis: Find the top 5 most popular works
top_works = df_filtered.nlargest(5, 'ratings_count')
print("\nTop 5 Most Popular Works:")
print(top_works[['original_title', 'ratings_count', 'avg_rating']])
# Example analysis: Count the number of works by genre
genre_counts = df_filtered['genres'].value_counts()
print("\nNumber of Works by Genre:")
print(genre_counts)

```

```python
#plot top 5 authors by number of works
import matplotlib.pyplot as plt
top_authors = author_counts.nlargest(5)
top_authors.plot(kind='bar', color='skyblue')
plt.title('Top 5 Authors by Number of Works')
plt.xlabel('Author')
plt.ylabel('Number of Works')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
